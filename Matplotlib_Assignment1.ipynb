```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
df = pd.DataFrame({
    'SMedia' : [
        'X' if i in range(1, 45+1) else 'Instagram' if i in range(45+1, 45+70+1) else 'Snapchat' for i in range(1, 145+1)
    ]
})

Platform = list(df['SMedia'].value_counts().index)
counts     = list(df['SMedia'].value_counts().values)
plt.bar(
    x = Platform,            
    height = counts,          
)
plt.show();
```

```python
data = [2, 4, 4, 5, 6, 6, 7, 7, 7, 8, 8, 8,9,9,10,11,11,11, 12, 13]

plt.hist(data, bins=10, edgecolor='black')
plt.title('Histogram of Data')
plt.xlabel('Data Values')
plt.ylabel('Frequency')

plt.show()

```

## Q2: Create pie chart with equal explode values.

```python
x = df['SMedia'].value_counts()
plt.pie(x,autopct='%1.1f%%', startangle=90,explode = [0.1,0.1,0.1])
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = np.random.randn(100)
y = np.random.randn(100)

plt.scatter(
    x, y, 
    alpha=0.4, marker='o', s=50
)
plt.title('Relationship Between Two Values', fontsize=15)
plt.xlabel('Value One', fontsize=12)
plt.ylabel('Value Two', fontsize=12)
plt.show();
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
data = [2, 4, 4, 5, 6, 6, 7, 7, 7, 8, 8, 8,9,9,10,11,11,11, 12, 13]

plt.hist(data, bins=10, edgecolor='black')
plt.title('Histogram of Data',loc="left",color='red')
plt.xlabel('Data Values', fontsize=16)
plt.ylabel('Frequency',fontsize=16)

plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
x=[1, 2, 3, 4, 5]
y=[1,2,3,5,1]
y_1=[4,3,2,5,1]
y_2=[3,5,4,2,1]

plt.plot(x , y, color='r',label='red line')
plt.plot(x , y_2, color='b',label='blue line')
plt.plot(x , y_1, color='black',label='black line')

plt.title('Three Line Chart with Random Values', fontsize=15)
plt.xlabel('Value 1', fontsize=12)
plt.ylabel('Value 2', fontsize=12)
plt.legend(loc='upper left')

```

## Q6: Create subplots with three charts and figsize=20,10.

```python
fig, axs = plt.subplots(nrows=1,ncols=3, figsize=(20, 10))

z=[100, 200, 300, 400, 500]
y11=[100,200,300,500,400]
y22=[400,100,500,300,200]
y33=[300,100,200,400,500]

axs[0].plot(z, y11, color='r')
axs[0].set_title('First Plot')

axs[1].plot(z, y22, color='b')
axs[1].set_title('Second Plot')

axs[2].plot(z, y33, color='g')
axs[2].set_title('Third Plot')

plt.show()
```
